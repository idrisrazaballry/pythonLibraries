#### any() and all()

#### all() --> it gives True if all conditons are True otherwise False

#### any() --> it gives True if any of one conditionis True otherwise False

```python
print((arr1<5).all() and (arr2>5).all())
```

```python
print((arr1<5).all() or (arr2>5).all())
```

```python
print((arr1>5).any() or (arr2<5).any())
```

```python
print((arr1<5).any() and (arr2>5).any())
```
