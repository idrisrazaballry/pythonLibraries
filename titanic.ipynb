```python
def comp_type(row):
    f=row['fam_cnt']
    p=row['no_of_psgr']
    if p==1:
        return 'Solo'
    elif f==0 and p>1:
        return 'Friends'
    else:
        return 'Family'
```

```python
df['Companion_type']=df.apply(comp_type, axis=1)
```

```python
Royal_clean = [t.strip().lower() for t in ['The Countees','Jonkheer','Don','Dona','Master','Sir']]
Common_clean= [t.strip().lower() for t in ['Miss', 'Mr','Mrs','Lady','Ms','Mme','Mlle']]
Military_clean= [t.strip().lower() for t in ['Major','Col','Capt']]

def social_status(row):
    c=str(row['Title']).strip().lower()
    if c in Royal_clean:
        return 'Royal'
    elif c in Common_clean:
        return 'Commom'
    elif c in Military_clean:
        return 'Military'
    else:
        return 'other_professional'
```

## BARPLOT

#### It is used for both categorical as well as numerical data

#### Categorical --> Group,     Numerical ---> Numbers

#### It uses both 1 variable as well 2 variable as input, So we not able conclude that whether it is univarient or bivarient data

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df, x='gender', y='fare', color='pink', errorbar=None, estimator='mean') # errorbar --> is used to remove the error bar which is present on the top of everybar
c.bar_label(c.containers[0], fmt='%.1f')                                                    # estimator --> is used to find the aggregate functions like min,max,, sum, mean
plt.title('Avg fare by gender wise')                                                      
plt.yticks(range(0,20,2))
plt.show()
```

### OBSERVATION

#### From above chart, we conclude that, the Female passengers are more in terms of mean as compare to the male average

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df, x='gender', y='fare', errorbar=None, estimator='sum') # errorbar --> is used to remove the error bar which is present on the top of everybar
c.bar_label(c.containers[0], fmt='%.1f')                                                    # estimator --> is used to find the aggregate functions like min,max,, sum, mean
plt.title('Sum fare by gender wise')                                                      
plt.yticks(range(0,14000,2000))
plt.show()
```

### OBSERVATION

#### From above graph we conclude that, males are paying more ass compare to female

```python
plt.figure(figsize=(5,6))
c=sns.barplot(data=df, x='age_category', y='fare',hue='gender', errorbar=None, estimator='mean')
for i in c.containers:
    c.bar_label(i, fmt='%.1f')
plt.yticks(range(0,40,5))
plt.legend(loc='best',ncol=2, title='Gender')      # ncol ---> it is used to adjust the legend's table labels   # legend ---> it is used to adjust the legend box where-ever we want
plt.title('Avg fare of pgrs according to age category')
plt.show()
```

### OBSERVATION

#### From above graph we conclude that, the females are paying more as compare to male in every age_category and average fare is incresing according to the age.

## SUBPLOT

#### Subplot nothing but dividing/plotting the page into multiple (any kind of plot) sections


```python
g=df.loc[df['Companion_type']=='Family']
```

```python
gf=g['gender'].value_counts()
gn=g['age_category'].value_counts()
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(6,4))
plt.suptitle('Companion type wise ctegorization')

axes[0].pie(x=gf.values, labels=gf.index, autopct='%.1f%%')
axes[0].set_title('Gender wise categorization')

axes[1].pie(x=gn.values, labels=gn.index, autopct='%.1f%%')
axes[1].set_title('Age wise categorization')

plt.tight_layout()                               # tight__layout ---> it is used to adjust the title and subtitles in the main graph
plt.show()
```

### OBSERVATION

#### From above graph we conclude that, in campanion type more females are travelling as commpare to the male while in age_category adults are more as compare to others.
