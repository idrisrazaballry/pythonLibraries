```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y_true = [1]*50 + [0]*50 # 50 actual spam, 50 actual not spam
y_pred = [1]*40 + [0]*10 + [1]*5 + [0]*45 # 40 TP, 10 FN, 5 FP, 45 TN
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)
# [[TN FP]
# [FN TP]] = [[45 5]
# [10 40]]
```

```python
# Extract values
tn, fp, fn, tp = cm.ravel()
```

```python
# 1. Accuracy
accuracy = accuracy_score(y_true, y_pred)
# or manually: (tp + tn) / (tp + tn + fp + fn)
print(f"Accuracy: {accuracy:.2f}") # 0.85
```

```python
# 2. Precision
precision = precision_score(y_true, y_pred)
# or manually: tp / (tp + fp)
print(f"Precision: {precision:.2f}") # 0.89
```

```python
# 3. Recall / Sensitivity
recall = recall_score(y_true, y_pred)
# or manually: tp / (tp + fn)
print(f"Recall: {recall:.2f}") # 0.80
```

```python
# 4. F1-Score
f1 = f1_score(y_true, y_pred)
# or manually: 2 * (precision * recall) / (precision + recall)
print(f"F1-Score: {f1:.2f}") # 0.84
```

```python
# 5. Specificity
specificity = tn / (tn + fp)
print(f"Specificity: {specificity:.2f}") # 0.90
```

```python
# 6. Print full classification report
from sklearn.metrics import classification_report
print("\nFull Report:")
print(classification_report(y_true, y_pred, target_names=['Not Spam', 'Spam']))
```

```python
# 7. Plot confusion matrix
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Spam', 'Spam'],
            yticklabels=['Not Spam', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
