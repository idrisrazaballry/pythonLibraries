```python
def job_clean(row):
    j=row['job']
    d=row['dName']
    if j!='Auditor' and j!='Dairy Person':
        if d in j:
           return j.replace(d,'').strip()
        else:
            return j.removeprefix('Exec Assistant').strip(', ')
    else:
        return j
```

```python
df.columns
```

```python
df['job']=df.apply(job_clean, axis=1)
```

```python
plt.figure(figsize=(3,5))
c=sns.countplot(data=df, x='status', color='black')
c.bar_label(c.containers[0])
plt.yticks(range(0, 8000, 1000))
plt.ylabel('No of emps')
plt.title('Distribution of emps according to status')
plt.show()
```

#### OBSERVATION:

##### From above graph we conclude that, number of active emps are more as compare to terminated emps

```python
val1=df['gender'].value_counts()
val1.values
val1.index
```

```python
plt.figure(figsize=(4,6), facecolor='silver')
plt.pie(x=val1.values, labels=val1.index, colors=['gold', 'green'], autopct='%.2f%%', pctdistance=0.7, labeldistance=1.1)
plt.title('Distribution of emp gender wise')
plt.show()
```

#### OBSERVATION

```python
plt.figure(figsize=(4,6))
cnt=sns.barplot(df, x='status', y='experience',hue='gender', errorbar=None)
for i in cnt.containers:
    cnt.bar_label(i, fmt='%.1f', padding=-17)
plt.yticks(range(0, 20, 2))
plt.legend(loc='best', ncol=2, title='Gender')
plt.title('Avg experince of emps gender wise')
plt.show()
```
