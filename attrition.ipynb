```python
def job_clean(row):
    j=row['job']
    d=row['dName']
    if j!='Auditor' and j!='Dairy Person':
        if d in j:
           return j.replace(d,'').strip()
        else:
            return j.removeprefix('Exec Assistant').strip(', ')
    else:
        return j
```

```python
df.columns
```

```python
df['job']=df.apply(job_clean, axis=1)
```

```python
h_df=df['hiredDate'].dt.year.value_counts().sort_values().reset_index()
```

```python
h_df.columns=['year', 'hiring']
```

```python
t_df=df.loc[df['status']=='Terminated', 'termDate'].dt.year.value_counts().sort_values().reset_index()
```

```python
t_df.columns=['year','termination']
```

```python
df1=pd.merge(h_df,t_df,on='year',how='outer').fillna(0).astype(int)
```

```python
df1['retention1']=df1['hiring']-df1['termination']
```

```python
df1['retention']=df1['retention1'].cumsum()       # cumsum --> cummulative sum
```

```python
df1=df1.drop(columns='retention1')
```
