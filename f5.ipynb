```python
import numpy as np
```

```python
a1=np.arange(10,96,5)
print(a1)
print(a1.shape)
```

### Reshape Functions

#### It is going to convert one shape of array into another.
#### If we want to reshape the array then we have to check with multiple of that shape.

```python
np.reshape(a1, shape=(6,3))
```

```python
np.reshape(a1,(3,3,2))
```

```python
a2=np.reshape(a1,(2,3,3))
print(a2)
```

-1 is used to convert any type of array into 1-D array

```python
np.reshape(a2,(-1))
```

```python
a3=np.array([[10,20,30,40,50,78],[2,4,5,60,64,28],[11,9,37,92,55,68]])
print(a3)
print(a3.shape)
print(a3.ndim)       # It is used to check the dimension of the array
```

```python
np.reshape(a3, (3,2,1,3))
np.reshape(a3,(6,3))
```
