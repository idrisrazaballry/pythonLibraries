```python
arr1=np.array([10,20,30])
print(arr1)
```

```python
s_a=np.copy(arr1) # it is used to create the shallow copy of particular array
print(s_a)

s_a[0]=110
print(s_a)
print()
print(arr1)
```

```python
arr2=arr1.view() # It is used to copy the array but it stores the elements in same location.
print('arr2:',arr2)
print('arr1:',arr1)
print()

arr2[-1]=200   # The main difference b\w copy and view method is, when we update the array with new elements, the copy() willn't update with original array but when come to view(), it also update with original array
print('arr2:',arr2)
print('arr1:',arr1)
```

```python
arr3=np.array([[1,2,3],[10,20,30]])
print(arr3)
```

```python
x=arr3.ravel() # It is used to convert any type of array into 1D array like reshape() but when we update the array with new element then it works as view(), it update the value with both original array as well as updated array
print(x)
print()

x[0]=50
print(x)
print()
print(arr3)
```

```python
arr4=np.array([[11,22,33],[4,67,80]])
print(arr4)
print()

y=arr4.flatten()   # It also works like reshape() method, which converts the any type of array into 1D arrray and it is independent
print(y)
print()

y[-1]=900       # When we update the array with new elements, it doesn't update with original array
print(y)
print(arr4)
```
