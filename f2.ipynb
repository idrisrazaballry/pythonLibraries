```python
import numpy.random as rd

rd.randint(10)        # randint --> it is used to generate the random integer numbers
```

```python
rd.randint(10,100)  # it is used to generate the random integer number between 10 to 100
```

```python
rd.randint(10,100,(3)) # It is used to create the random number in form of matrix
```

```python
rd.randint(10,100,(3,2)) # It is used to create the random number in form of matrix and here 3 indicates the number of rows and 2 indicates number of columns
```

```python

rd.seed(5)  # It is used to fix the number
rd.randint(10,100,(3,2))
```

```python
rd.rand() # It is used to genearte the folating point number between 0 & 1 and that number must be positive number
```

```python
rd.rand(3)
```

```python
rd.rand(2,3)
```

```python
''' randn() method 
It is same is rand() but randn() generates the negative numbers also and it is in the range between -1 & 1
Its mean is near to zero (means it is in negative form) and standard deviation is near to one (means it is in positive form)
'''

r=rd.randn(2)
print(r)

print(np.mean(r))
print(np.std(r))
```
